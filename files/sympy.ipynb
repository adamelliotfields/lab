Plotting algebraic functions with [SymPy](https://www.sympy.org/en/index.html).

```python
import matplotlib.pyplot as plt
import sympy as sp
```

## Linear Functions

$y = mx + b$

```python
x = sp.symbols("x")
m = 2
b = 1
y = m * x + b
sp.plot(y, (x, -5, 5), title="$y=2x+1$")
plt.show()
```

## Quadratic Functions

$y = ax^2 + bx + c$

```python
x = sp.symbols("x")
a = 1
b = -4
c = 3
y = a * x**2 + b * x + c
sp.plot(y, (x, -5, 5), title="$y=x^2-4x+3$")
plt.show()
```

## Cubic Functions

$y = ax^3 + bx^2 + cx + d$

```python
x = sp.symbols("x")
a = 1
b = -3
c = 2
d = 0
y = a * x**3 + b * x**2 + c * x + d
sp.plot(y, (x, -5, 5), title="$y = 1x^3 + -3x^2 + 2x + 0$")
plt.show()
```

## Exponential Functions

$y = ab^x$

```python
x = sp.symbols("x")
a = 1
b = 2
y = a * b**x
sp.plot(y, (x, -5, 5), title="$y = 1 \cdot 2^x$")
plt.show()
```

## Logarithmic Functions

$y = a * \log_b(x) + c$

```python
x = sp.symbols("x")
a = 1
b = 2
c = 1
y = a * sp.log(x, b) + c
sp.plot(y, (x, -5, 5), title="$y = 1 \cdot log_2(x) + 1$")
plt.show()
```

# Absolute Value Functions

$y = |x|$

```python
x = sp.symbols("x")
y = sp.Abs(x)
sp.plot(y, (x, -5, 5), title="$y = |x|$")
plt.show()
```

## Rational Functions

$y = \frac{ax + b}{cx + d}$

```python
x = sp.symbols("x")
a = 1
b = 1
c = 1
d = -1
y = (a * x + b) / (c * x + d)
sp.plot(y, (x, -10, 10), title="$y = (1x + 1) / (1x - 1)$", ylim=(-10, 10))
plt.show()
```

## Logistic (Sigmoid) Functions

$y = \frac{1}{1 + e^{-x}}$

```python
x = sp.symbols("x")
y = 1 / (1 + sp.exp(-x))
sp.plot(y, (x, -5, 5), title="$y = 1 / 1 + e^-x")
plt.show()
```

## Piecewise Functions

$y = \begin{cases}
  f_1(x) & \text{if } x < a \\
  f_2(x) & \text{if } x \geq a
\end{cases}$

```python
x = sp.symbols("x")
y = sp.Piecewise((x**2, x < 0), (sp.sqrt(x), x >= 0))
sp.plot(y, (x, -5, 5), title="$y = x^2$ if $x < 0$, $y = \sqrt{x}$ if $x \geq 0$", ylim=(-10, 10))
plt.show()
```
